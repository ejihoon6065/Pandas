```python
import pandas as pd
import numpy as np
```

```python
# 2차원 리스트 생성
# 세로로 시리즈가 형성됨
# 보이는대로 출력이 된다
data1 = [
    [10,20,30,40,50],
    [100,200,300,400,500],
    [1000,2000,3000,4000,5000]    
]
col1 = ['영어', '수학', '사회', '과학', '국어']
index1 = ['v1', 'v2', 'v3']
df1 = pd.DataFrame(data1, columns=col1, index = index1)
df1

# numpy행렬 생성도 가능
```

```python
# 딕셔너리로 데이터 프레임 생성
# 가로로 시리즈가 형성됨
# 딕셔너리의 인덱스가 각각 열로 삽입됨
data2 = {
    'a1' : [10,100,1000],
    'a2' : [20,200,2000],
    'a3' : [30,300,3000],
    'a4' : [40,400,4000], 
    'a5' : [50,500,5000], 
    }

index2 = ['v1', 'v2', 'v3']
df2 = pd.DataFrame(data2, index = index2)
df2
```

```python
# 파일에서 데이터 읽어오기
df3 = pd.read_csv('iris.csv')
df3
```

```python
# info() : 데이터 프레임의 간략한 정보를 확인할 수 있다.
# RangeIndex에서 전체 row의 개수를 확인할 수 있고, 각 컬럼별 row의 개수는 결측치를 제외한 개수를 표시해줌
# 이는 각 컬럼에서 결측치가 존재하는지 여부를 확인하게 해줌

df3.info()
#150 entries가 존재하기 때문에 개별 column의  non-null count가 150개라는 것은 결측치가 없다는 것
```

```python
# head() : 상위 n개의 데이터를 추출하고 새로운 데이터 프레임으로 만들어준다. n을 생략할 경우 5개를 추출.
df3.head(7)
```

```python
#tail(): 하위 n개의 데이터를 추출하고 새로운 데이터 프레임으로 만들어준다. n을 생략할 경우 5개를 추출.
df3.tail(7)
```
